```python
crop_dict = {}
for col in  crop_matrix.columns:
    val = ()
    crop_dict.update({col:sorted([(index, crop_matrix[col][index]) for index in crop_matrix.index],
    key = lambda elem: elem[1])})
crop_matrix
```

```python
import networkx as nx

G = nx.Graph()
options = {
    'node_color': 'black',
    'edge_color': 'red',
    'node_size': 50,
    'width': 3,
}
```

```python
for key in crop_dict.keys():
    for elem in crop_dict[key]:
        G.add_edge(key, elem[0], weight=elem[1])
```

```python
nx.draw_circular(G, **options, with_labels = True)
plt.show()
```
